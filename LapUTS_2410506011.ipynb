```python
#mendeteksi jumlah bilangan ganjil dan genap pada NPM masing masing mahasiswa
npm = input("Masukkan NPM Anda")

genap = 0
ganjil = 0

for angka in npm:
    if int(angka) % 2 == 0:
        genap += 1
    else:
        ganjil += 1

print("Jumlah angka genap:", genap)
print("Jumlah angka ganjil:", ganjil)
```
